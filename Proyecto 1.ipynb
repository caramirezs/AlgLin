# **Taller 1: USS Orthon Perdida en la Dimensión Desconocida**  

## **Parte 1: Rectificando la Dimensión Perdida**  
En una misión de exploración lejana, el **Capitán Zethron** y su valiente tripulación a bordo de la nave **USS Orthon** se encuentran atrapados en lo que parece ser... **otra dimensión**.  

— **¡Capitán!**, algo no está bien... — exclama la piloto.  
— **¡Reporten la situación!** — ordena el Capitán Zethron, con una calma que apenas oculta su preocupación.  
— Es extraño... el espacio aquí parece **bidimensional**, pero todo está **distorsionado**. ¡Como si la base de esta dimensión no fuera ortonormal!  

Tras unos minutos de análisis, el sistema de sensores de la **USS Orthon** logra captar datos de lo que parece ser otra nave a lo lejos. Sin embargo, la imagen está deformada, completamente fuera de proporción.  

— **¡Rápido!** — ordena el capitán — ¿Alguien puede transformar estos datos a una base ortonormal para ver la nave como realmente es?  
— **¡Por supuesto, Capitán!** — respondes con confianza. **Este es exactamente el tipo de problema que mi equipo está entrenado para resolver.**

## **Ejercicio: Rectificando la Dimensión Perdida**  

### **1. Descarga los datos proporcionados por el piloto.**  
Hemos recibido los datos de la posición de puntos clave de la nave en esta dimensión distorsionada. Descarga el archivo CSV con las coordenadas.

### **2. Análisis de la Base**  
Después de un análisis riguroso utilizando **regresión espectral bidimensional**, concluyen que la base de esta dimensión está dada por los vectores:  

$$\mathbf{b_1} = (1.9, 1.2) \quad \text{y} \quad \mathbf{b_2} = (1.2, 2.4)$$

**Tarea:**  
Representa en el mismo plano los datos de la nave y esta base. ¡Es cierto! **La nave se ve distorsionada.**

### **3. Corrige la Base**  
Es hora de trabajar. Utiliza el proceso de **Gram-Schmidt** para convertir la base dada en una base ortonormal. Representa nuevamente la nave utilizando esta base ortonormal y observa cómo se rectifica la distorsión.

## **Parte 2: Misión de Ilusión Estratégica – Operación *Phantom Fleet***  

El ambiente en la sala de comando es tenso. Los oficiales debaten sobre qué hacer a continuación.  
**¿Contactar a la nave desconocida? ¿Atacar? ¿Pedir refuerzos?**  

— **¡No! Los refuerzos no llegarían a tiempo... podríamos desatar una guerra interdimensional!** — murmura uno de los oficiales.  

De repente, se te ocurre una idea. **"Capitán, podemos engañar los sensores de la nave enemiga."**  

— **¿Engañar los sensores? ¿Cómo?**  

Explicas tu plan:  

— *"Ya que tenemos los datos y las características básicas de la nave, podemos usar **transformaciones lineales** para simular una flota de naves aliadas: generar réplicas de la nave en diferentes posiciones, con rotaciones y reflejos. Así creerán que somos una flota completa y probablemente se retirarán."*  

Un silencio tenso invade la sala. El Capitán Zethron lo rompe:  

— **No me convence... ¿Alguna otra idea?**  

El silencio se prolonga unos segundos más hasta que el Capitán finalmente dice:  
— **Ugh... no hay más tiempo. Procedan con la misión... ¡Operación *Phantom Fleet*!**

**Objetivo:** Utilizar transformaciones lineales para crear una flota de **8 naves** diferentes, cada una con una combinación única de:

## Epílogo
Tu equipo transmite rápidamente los datos a la USS Orthon, que los canaliza a través del sistema de proyección holométrica de la nave. Inmediatamente, un gran holorama de dos dimensiones disfraza la Orthon como una flota de ocho naves idénticas, desplegadas en el espacio bidimensional.

La tensión en la sala de mando es palpable. La luz del puente parpadea ligeramente, el zumbido del reactor suena más fuerte... hasta parece que el tiempo mismo se ralentiza.

— ¡Capitán, qué hacemos, atacamos! — grita el oficial de tácticas, con el dedo firme en el gatillo, a punto de presionar el botón de disparo.

— No. Esperemos... — responde el Capitán Zethron, con mirada penetrante.

De repente, la nave enemiga gira lentamente y cambia su rumbo. Se aleja con cautela... más y más lejos, hasta que es solo un pequeño punto en la pantalla.

— ¿Cómo es posible? ¡Estamos en un espacio bidimensional! — pregunta alguien en voz baja.

— Eso lo dejamos para los científicos. — responde el Capitán Zethron, esbozando una sonrisa. — Por lo pronto... salgamos de aquí.

La USS Orthon ajusta sus motores y comienza a deslizarse suavemente hacia el hiperumbral, dejando atrás aquella dimensión extraña y distorsionada, mientras el equipo respira aliviado tras una misión bien cumplida.
